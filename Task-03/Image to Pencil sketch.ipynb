## Importing all the libraries

```python
#for plotting
import matplotlib.pyplot as plt

#for reading and manipulating images
import cv2

```

### Reading image using opencv library and exploring more knowledge about them

#### OpenCV is a cross-platform library using which we can develop real-time computer vision applications. 
#### It mainly focuses on image processing.
#### video capture and analysis including features like face detection and object detection.

```python
img = cv2.imread('Aashish.jpg', 1)

img.shape
```

#### Why there's lot of numbers where we read image !!!

##### Computer doesn't recognise a image as an image. So, we had to go differently. Computer screen are full of pixels. So, we computer create image by the combination of only three colors RGB ( Red , Green and Blue ). Each pixels has only these three colors . If we mix RGB with varying density we can create whole lot of different colors. Density of each color vary from 0 to 255. Similarly, open cv also keeps images in the array of numbers where each number denotes the intensity of Blue , Red and Green (BGR) color on the respective pixel.

```python
plt.imshow(img)
```

```python
#Coverting BGR to RGB

img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

#### To decrease the complexity of the image and make the handling easier we will be converting the image to a grayscale image.

```python
#RGB to GrayScale

img_gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray)
```

### Now we are going invert this grayscale image which will help me to understand or highlight it in more detail.

```python
img_invert = cv2.bitwise_not(img_gray)
plt.imshow(img_invert)

# To make the image less with edge i,e more edge we are going to blur it using Gaussin_Blur
blur_img = cv2.GaussianBlur(img_invert, (21, 21), 0) #
plt.title("Blurred Image")
```

## Inverting the blur

```python
invert_blur_img = cv2.bitwise_not(blur_img)

plt.imshow(invert_blur_img)
plt.title("Inverted Blurred Image")
```

```python
sketch_img = cv2.divide(img_gray, invert_blur_img, scale=250.0)

## Visualizing image using matplotlib

plt.imshow(sketch_img)
plt.axis('off')
plt.title("Pencil Sketched Image").set_color("purple")
```

```python
cv2.imwrite('./Aashish1.jpg', sketch_img)
```
